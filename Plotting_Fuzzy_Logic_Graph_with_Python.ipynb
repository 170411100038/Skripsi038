**Import library yang dibutuhkan**

```python
import numpy as np
# %pip install scikit-fuzzy
import skfuzzy as fuzz
import matplotlib.pyplot as plt  
```

**It's time to plotting!**

```python
def fuzzy_preg():
    x = np.arange(0, 20, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 1, 2])  
    normal = fuzz.trapmf(x,[1, 2, 4, 5])  
    tinggi = fuzz.trapmf(x,[4, 5, 20, 20])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Pregnancies')  
    ax0.legend()
    
def fuzzy_glucose():
    x = np.arange(0, 250, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 135, 145])  
    normal = fuzz.trapmf(x,[135, 145, 195, 205])  
    tinggi = fuzz.trapmf(x,[195, 205, 250, 250])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Glucose')  
    ax0.legend()
    
def fuzzy_bp():
    x = np.arange(0, 200, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 55, 65])  
    normal = fuzz.trapmf(x,[55, 65, 75, 85])  
    tinggi = fuzz.trapmf(x,[75, 85, 200, 200])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Blood Pressure')  
    ax0.legend()
    
def fuzzy_sk():
    x = np.arange(0, 70, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 15.5, 17.5])  
    normal = fuzz.trapmf(x,[15.5, 17.5, 28, 30])  
    tinggi = fuzz.trapmf(x,[28, 30, 150, 150])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Skin Thickness')  
    ax0.legend()

def fuzzy_insulin():
    x = np.arange(0, 200, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 115, 125])  
    normal = fuzz.trapmf(x,[115, 125, 135, 145])  
    tinggi = fuzz.trapmf(x,[135, 145, 200, 200])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Insulin')  
    ax0.legend()

def fuzzy_bmi():
    x = np.arange(0, 50, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 16, 21])  
    normal = fuzz.trapmf(x,[16, 21, 22.5, 27.5])  
    tinggi = fuzz.trapmf(x,[22.5, 27.5, 100, 100])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks BMI')  
    ax0.legend()
    
def fuzzy_dpf():
    x = np.arange(0, 3, 0.1)
    
    rendah = fuzz.trapmf(x,[0, 0, 0.4, 0.6])  
    normal = fuzz.trapmf(x,[0.4, 0.6, 0.9, 1.1])  
    tinggi = fuzz.trapmf(x,[0.9, 1.1, 3, 3])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Rendah')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Normal')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tinggi')
    ax0.set_title('Indeks Diabetes Predigree Function')  
    ax0.legend()

def fuzzy_age():
    x = np.arange(0, 50, 1)
    
    rendah = fuzz.trapmf(x,[0, 0, 15, 19])  
    normal = fuzz.trapmf(x,[15, 19, 22, 26])  
    tinggi = fuzz.trapmf(x,[22, 26, 100, 100])  
    
    fig, (ax0) = plt.subplots(nrows=1, figsize=(6, 4))

    ax0.plot(x, rendah, 'b', linewidth=1.5, label='Muda')  
    ax0.plot(x, normal, 'g', linewidth=1.5, label='Dewasa')  
    ax0.plot(x, tinggi, 'r', linewidth=1.5, label='Tua')
    ax0.set_title('Indeks Age')  
    ax0.legend()

fuzzy_preg()
fuzzy_glucose()
fuzzy_bp()
fuzzy_sk()
fuzzy_insulin()
fuzzy_bmi()
fuzzy_dpf()
fuzzy_age()
```
