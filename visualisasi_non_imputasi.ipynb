```python
import pandas as pd
import numpy as np
```

```python
def Average(dat):
    return sum(dat) / len(dat)
```

```python
#Data Output ID3
dat_id3 = pd.read_excel('output_ID3_nonImput_2.xlsx')
akurasi_id3 = dat_id3["Akurasi"]
presisi_id3 = dat_id3["Presisi"]
recall_id3 = dat_id3["Recall"]
ratAkurasi_ID3 = round(Average(akurasi_id3), 2)
ratPresisi_ID3 = round(Average(presisi_id3), 2)
ratRecall_ID3 = round(Average(recall_id3), 2)

#Data Output C45
dat_c45 = pd.read_excel('output_C45_nonImput_2.xlsx')
akurasi_c45 = dat_c45["Akurasi"]
presisi_c45 = dat_c45["Presisi"]
recall_c45 = dat_c45["Recall"]
ratAkurasi_C45 = round(Average(akurasi_c45), 2)
ratPresisi_C45 = round(Average(presisi_c45), 2)
ratRecall_C45 = round(Average(recall_c45), 2)

#Data Output ID3 Bagging
ID3_bagAk = pd.read_excel("ID3_bagging_nonImput_akurasi_2.xlsx")
ID3_bagPr = pd.read_excel("ID3_bagging_nonImput_recall_2.xlsx")
ID3_bagRe = pd.read_excel("ID3_bagging_nonImput_presisi_2.xlsx")
ID3_bag_hasil_akhir = pd.read_excel("ID3_bagging_nonImput_hasil_akhir_2.xlsx")
ID3_ratAkurasi_bagging = [ID3_bag_hasil_akhir['k:'+str(i)][0] for i in  range(2,ID3_bag_hasil_akhir.shape[1]+1)]
ID3_ratPresisi_bagging = [ID3_bag_hasil_akhir['k:'+str(i)][1] for i in  range(2,ID3_bag_hasil_akhir.shape[1]+1)]
ID3_ratRecall_bagging = [ID3_bag_hasil_akhir['k:'+str(i)][2] for i in  range(2,ID3_bag_hasil_akhir.shape[1]+1)]
ID3_maxAkurasi_Bagging = max(ID3_ratAkurasi_bagging)
ID3_maxPresisi_Bagging = max(ID3_ratPresisi_bagging)
ID3_maxRecall_Bagging = max(ID3_ratRecall_bagging)

#Data Output C45 Bagging
C45_bagAk = pd.read_excel("C45_bagging_nonimputasi_akurasi_2.xlsx")
C45_bagPr = pd.read_excel("C45_bagging_nonimputasi_recall_2.xlsx")
C45_bagRe = pd.read_excel("C45_bagging_nonimputasi_presisi_2.xlsx")
C45_bag_hasil_akhir = pd.read_excel("C45_bagging_nonimputasi_hasil_akhir_2.xlsx")
C45_ratAkurasi_bagging = [C45_bag_hasil_akhir['k:'+str(i)][0] for i in  range(2,C45_bag_hasil_akhir.shape[1]+1)]
C45_ratPresisi_bagging = [C45_bag_hasil_akhir['k:'+str(i)][1] for i in  range(2,C45_bag_hasil_akhir.shape[1]+1)]
C45_ratRecall_bagging = [C45_bag_hasil_akhir['k:'+str(i)][2] for i in  range(2,C45_bag_hasil_akhir.shape[1]+1)]
C45_maxAkurasi_Bagging = max(C45_ratAkurasi_bagging)
C45_maxPresisi_Bagging = max(C45_ratPresisi_bagging)
C45_maxRecall_Bagging = max(C45_ratRecall_bagging)


print('Data ID3')
print(dat_id3,'\n')
print('Data C45')
print(dat_c45,'\n')
print('Data ID3 Bagging')
print(ID3_bag_hasil_akhir)
print('Data C45 Bagging')
print(C45_bag_hasil_akhir)
```

```python
def tabelPerbandinganMetode():
    dat = []

    dat.append(["Akurasi",ratAkurasi_ID3, ID3_maxAkurasi_Bagging, ratAkurasi_C45, C45_maxAkurasi_Bagging])
    dat.append(["Presisi",ratPresisi_ID3, ID3_maxPresisi_Bagging, ratPresisi_C45, C45_maxPresisi_Bagging])
    dat.append(["Recall",ratRecall_ID3, ID3_maxRecall_Bagging, ratRecall_C45, C45_maxRecall_Bagging])
    
    dat2 = pd.DataFrame(dat, columns=['Keterangan', 'ID3', 'ID3 Bagging', 'C45', 'C45 Bagging'])

    return dat2

PerbandinganMetode = tabelPerbandinganMetode()
print(PerbandinganMetode)
```

```python
import matplotlib.pyplot as plt

k = [('k:'+str(i)) for i in range(2,16)]

def ID3_bagging_akurasi() :
    plt.plot(k, ID3_ratAkurasi_bagging, label='Akurasi', marker='o',  markersize=5)

    plt.xlabel('k Bagging')
    plt.ylabel('Akurasi')
    plt.title('Perbandingan akurasi ID3 Bagging')

    plt.grid()
    plt.show()

def ID3_bagging_presisi() :
    plt.plot(k, ID3_ratPresisi_bagging, label='Presisi', marker='o',  markersize=5)

    plt.xlabel('k Bagging')
    plt.ylabel('Presisi')
    plt.title('Perbandingan presisi ID3 Bagging')

    plt.grid()
    plt.show()

def ID3_bagging_recall() :
    plt.plot(k, ID3_ratRecall_bagging, label='Recall', marker='o',  markersize=5)

    plt.xlabel('k Bagging')
    plt.ylabel('Recall')
    plt.title('Perbandingan recall ID3 Bagging')

    plt.grid()
    plt.show()

ID3_bagging_akurasi()
ID3_bagging_presisi()
ID3_bagging_recall()
```

```python
def perbandingan_metode_id3():
    # set width of bar
    barWidth = 0.30
    fig = plt.subplots(figsize =(12, 8))

    # set height of bar
    label = ['ID3', 'ID3 Bagging']
    # Akurasi = [PerbandinganMetode[label[i]][0] for i in range(len(label))]
    ID3_Single = PerbandinganMetode['ID3']
    ID3_Bagging = PerbandinganMetode['ID3 Bagging']

    # Set position of bar on X axis
    br1 = np.arange(len(ID3_Single))
    br2 = [x + barWidth for x in br1]

    # Make the plot
    plt.bar(br1, ID3_Single, color ='tomato', width = barWidth,edgecolor ='grey', label ='ID3')
    plt.bar(br2, ID3_Bagging, color ='skyblue', width = barWidth,edgecolor ='grey', label ='ID3 Bagging')

    # Adding Xticks
    plt.xlabel('Evaluasi Hasil', fontweight ='bold', fontsize = 15)
    plt.ylabel('Persentasi', fontweight ='bold', fontsize = 15)
    plt.xticks([r + barWidth for r in range(len(ID3_Single))],['Akurasi', 'Presisi', 'Recall'])

    plt.legend()
    plt.show()
    
perbandingan_metode_id3()
```
